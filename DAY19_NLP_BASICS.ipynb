```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

print("Ready for NLP!!")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv',
                 sep='\t', header=None, names=['label', 'message'])

print(df.shape)
print(df.head())
print("\nLabel counts:")
print(df['label'].value_counts())
```

```python
# Convert labels to numbers
df['label_num'] = (df['label'] == 'spam').astype(int)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label_num'],
    test_size=0.2, random_state=42)

# Convert text to numbers using Bag of Words!!
vectorizer = CountVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

print("Vocabulary size:", len(vectorizer.vocabulary_))
print("Training matrix shape:", X_train_vec.shape)
```

```python
# Train Naive Bayes model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Predict
predictions = model.predict(X_test_vec)

# Results
print(f"Accuracy: {accuracy_score(y_test, predictions)*100:.2f}%")
print("\n", classification_report(y_test, predictions,
      target_names=['ham', 'spam']))
```

```python
# Test with custom messages!!
test_messages = [
    "Hey Freebie, how are you today??",
    "FREE moisturizer with every order , use this code",
    "Congratulations on your research work , brother can wee meet at coffee..i am emailing you ..heheh??",
    "Congratulations , you are accepted!!"
]

test_vec = vectorizer.transform(test_messages)
results = model.predict(test_vec)

for msg, result in zip(test_messages, results):
    label = "SPAM 🚨" if result == 1 else "HAM ✅"
    print(f"{label}: {msg[:50]}")
```
