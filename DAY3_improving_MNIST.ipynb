```python
# === STEP 1: RELOAD DATA FROM SCRATCH ===
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

# Load fresh data
(train_images, train_labels), (test_images, test_labels) = keras.datasets.mnist.load_data()

print("Original shapes:")
print(f"train_images: {train_images.shape}")
print(f"train_labels: {train_labels.shape}")

# === STEP 2: NORMALIZE ===
train_images = train_images / 255.0
test_images = test_images / 255.0

print("\nAfter normalization:")
print(f"train_images min: {train_images.min()}, max: {train_images.max()}")
print(f"test_images min: {test_images.min()}, max: {test_images.max()}")

# === STEP 3: BUILD MODEL ===
model_improved = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10, activation='softmax')
])

# === STEP 4: COMPILE ===
model_improved.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# === STEP 5: TRAIN ===
history = model_improved.fit(
    train_images,
    train_labels,
    epochs=10,
    validation_split=0.2,
    verbose=1
)

# === STEP 6: EVALUATE ===
test_loss, test_accuracy = model_improved.evaluate(test_images, test_labels)
print(f"\nğŸ¯ Final Test Accuracy: {test_accuracy*100:.2f}%")
```

In deep learning, training loss and validation loss are key metrics for evaluating how well a model is learning and generalizing.

#Training loss :

measures the modelâ€™s error on the training dataset. It is computed after each forward and backward pass using a chosen loss function, such as cross-entropy for classification or mean squared error for regression. A lower training loss generally means the model is fitting the training data well.

#Validation loss:

 measures the modelâ€™s error on unseen validation data after each epoch. This helps assess generalization and detect overfitting. Unlike training loss, it is not used to update model weights.

Typical patterns and interpretations:

#Both losses decrease â†’ Healthy learning and good generalization.

#Training loss â†“, validation loss â†‘ â†’ Overfitting; the model memorizes training data.

#Both losses remain high â†’ Underfitting;

 the model is too simple or learning is ineffective.

#Validation loss < training loss â†’
Possible due to regularization or easier validation samples.

#Sudden spikes in validation loss â†’ Could be noisy data or high learning rate.
