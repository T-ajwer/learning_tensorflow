```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/gabbygab1233/Crop-Recommender/main/Crop_recommendation.csv')

print("Ready!!")
print(df.shape)

```

```python
plt.figure(figsize = (15, 8))
sns.scatterplot(data = df , x = "rainfall" , y = "humidity" , hue = "label" , alpha = 0.5)
plt.title("Scatter Plot for relationship of Rainfall and humidity")
plt.show()
```

```python
plt.figure(figsize = (15 , 8))
sns.scatterplot(data = df , x ="temperature" , y = "rainfall" , hue ="label" , alpha = 0.5)
plt.title("Scatter plot for Relationship between Temperature  andRainfall")
plt.show()
```

```python
pakistani_climate = df[
    (df['temperature'] >=20) &(df['temperature']<=35) &
    (df['humidity' ] >= 50) &(df['humidity']<= 90)&
    (df['rainfall' ]>= 50)& (df['rainfall'] <=220)
]
print("Crops that thrive in Pakistani enviroments")
print(pakistani_climate['label'].value_counts())
```
