```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
plt.figure(figsize = (8,5))
df['Survived'].value_counts().plot(kind = "bar" , color = "red" and "green")
plt.title("Survived Titanic People")
plt.xlabel("Survived (0=No , 1=Yes)")
plt.ylabel("No of passengers")
plt.xticks(rotation = 0)
plt.show()
```

```python
from pandas.core import groupby
plt.figure(figsize= (8, 5))
df.groupby("Pclass")["Survived"].mean().plot(kind = "bar" , color = ["red", "green"])
plt.title("Survived People For Each Pclass")
plt.xlabel("Pclasses 1 , 2 , 3")
plt.ylabel("Survival Rate ")
plt.xticks(rotation = 0)
plt.show()
```
