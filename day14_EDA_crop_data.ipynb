```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/gabbygab1233/Crop-Recommender/main/Crop_recommendation.csv')
print(df.shape)
print(df['label'].value_counts())
```

```python
print("=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== BASIC STATS ===")
df.describe()
```

```python
print(df['label'].unique())
```

```python
plt.figure(figsize = (15 , 8))
df.groupby("label")["N"].mean().plot(kind = 'bar', color = ("red" or "green"))
plt.xlabel("Label")
plt.ylabel("nitrogen ")
plt.show()
```

```python
plt.figure(figsize = (15 , 8))
df.groupby("label")["P"].mean().plot(kind = 'bar', color = ("red" or "green"))
plt.xlabel("Label/ types of Crops")
plt.ylabel("Phosphorus Average Usage ")
plt.show()
```

```python
plt.figure(figsize = (15 , 8))
df.groupby("label")["K"].mean().plot(kind = 'bar')
plt.xlabel("Label")
plt.ylabel("Potassium  Uasage avergare  ")
plt.show()
```

```python
my_crops = ['rice', 'cotton', 'maize', 'mango', 'orange']
df_my = df[df['label'].isin(my_crops)]

df_my.groupby('label')[['N','P','K']].mean().plot(kind='bar', figsize=(12,6))
plt.title("NPK Requirements - Tajwer's Family Crops!!")
plt.ylabel("Average Nutrient Requirement")
plt.xticks(rotation=0)
plt.show()
```
